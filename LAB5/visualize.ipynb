## library import

```python
import matplotlib.pyplot as plt
import csv
import numpy as np
```

## data extraction from CSV files

```python
noise, sep, nsep = [], [], []

# open and format data 
with open('data_noise_2D.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        noise.append([float(row["x0"]),float(row["x1"]),float(row["x2"]),float(row["y"])])
with open('data_sep_2D.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        sep.append([float(row["x0"]),float(row["x1"]),float(row["x2"]),float(row["y"])])
with open('data_nsep_2D.csv', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    for row in csv_reader:
        nsep.append([float(row["x0"]),float(row["x1"]),float(row["x2"]),float(row["y"])])

noise, sep, nsep = np.asarray(noise), np.asarray(sep), np.asarray(nsep)
```

## Data visualization

```python
# Extract the x and y coordinates and class labels from the data
x1_noise = noise[:, 1]
x2_noise = noise[:, 2]
y_noise = noise[:, 3]

x1_sep = sep[:, 1]
x2_sep = sep[:, 2]
y_sep = sep[:, 3]

x1_nsep = nsep[:, 1]
x2_nsep = nsep[:, 2]
y_nsep = nsep[:, 3]

# Set the colors for each class
colors = ['red', 'green', 'blue']

# Create a scatter plot
fig, (ax3, ax1, ax2) = plt.subplots(1, 3, figsize = (15,5))
ax1.scatter(x1_noise, x2_noise, c=y_noise,s=50, cmap='plasma')
ax2.scatter(x1_nsep, x2_nsep, c=y_nsep,s=50, cmap='plasma')
ax3.scatter(x1_sep, x2_sep, c=y_sep,s=50, cmap='plasma')

# Show the plot
plt.show()
```
